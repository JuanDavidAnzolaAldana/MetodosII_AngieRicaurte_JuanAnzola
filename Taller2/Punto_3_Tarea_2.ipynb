```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def f(x,y):
    return ((x**4)*(y**2)-2*(x**2)*y-1)/(x**3)
```

```python
def Solucion(x):
    return (1/(x**2)) + 2/(-2 - x**2)
```

```python
N = 10
t = np.linspace((2**0.5),1,N)
```

```python
h = t[1] - t[0]
```

```python
yext = Solucion(t)
```

```python
plt.scatter(t,yext)
```

```python
def GetEuler(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1
```

```python
y1 = np.zeros_like(t)
y1[0] = 0
```

```python
GetEuler(t,y1,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1)
```

```python
def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
```

```python
GetEuler2(t,y1,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y1,label='Euler2')
plt.legend()
```

```python
def GetEuler3(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
        k3 = f(x[i-1]+h,y[i-1] - h*k1 + 2*h*k2)
        
        y[i] = y[i-1] + (1/6)*h*(k1+4*k2+k3)
```

```python
GetEuler3(t,y1,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y1,label='Euler2')
plt.plot(t,y1,label='Euler3')
plt.legend()
```

```python
def GetEuler4(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
        k3 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k2)
        k4 = f(x[i-1]+h,y[i-1] + h*k3)
        
        y[i] = y[i-1] + (1/6)*h*(k1+2*k2+2*k3+k4)
```

```python
GetEuler4(t,y1,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y1,label='Euler2')
plt.plot(t,y1,label='Euler3')
plt.plot(t,y1,label='Euler4')
plt.legend()
```
