```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
q = 0.

def f(x,y):
    return y**(q)

def Solucion(x,q):
    if q==1:
        return np.exp(x)
    else:
        return (x*(1-q)+1)**(1/(1-q))
    
N = 1000
t = np.linspace(0,50,N)
h = t[1] - t[0]
yext = Solucion(t,q)

def GetEuler(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1
        
y1 = np.zeros_like(t)
y1[0] = 0
GetEuler(t,y1,h)

def GetEuler2(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+h,y[i-1] + h*k1)
        
        y[i] = y[i-1] + 0.5*h*(k1+k2)
        
GetEuler2(t,y1,h)

def GetEuler3(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
        k3 = f(x[i-1]+h,y[i-1] - h*k1 + 2*h*k2)
        
        y[i] = y[i-1] + (1/6)*h*(k1+4*k2+k3)
        
GetEuler3(t,y1,h)

def GetEuler4(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k1)
        k3 = f(x[i-1]+0.5*h,y[i-1] + 0.5*h*k2)
        k4 = f(x[i-1]+h,y[i-1] + h*k3)
        
        y[i] = y[i-1] + (1/6)*h*(k1+2*k2+2*k3+k4)
        
GetEuler4(t,y1,h)

plt.title("q=0")
plt.plot(t,yext)
plt.plot(t,y1,label='Euler1')
plt.plot(t,y1,label='Euler2')
plt.plot(t,y1,label='Euler3')
plt.plot(t,y1,label='Euler4')
plt.legend()
```
