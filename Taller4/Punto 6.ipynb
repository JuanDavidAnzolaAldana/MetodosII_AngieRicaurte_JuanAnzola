Por la fórmula de Taylor se tiene que todoa función contnua se puede escribir como una serie de potencias:
$$y(t+\Delta t)=y(t)+y'(t)\Delta t+y''(t)\frac{\Delta t^2}{2}+y'''(t)\frac{\Delta t^3}{6}+y''''(t)\frac{\Delta t^4}{24}+y'''''(t)\frac{\Delta t^5}{120}+\mathcal{O}(\Delta t^6)$$
Si se cambia $\Delta t$ por $-\Delta t$ queda lo siguiente:
$$y(t-\Delta t)=y(t)-y'(t)\Delta t+y''(t)\frac{\Delta t^2}{2}-y'''(t)\frac{\Delta t^3}{6}+y''''(t)\frac{\Delta t^4}{24}-y'''''(t)\frac{\Delta t^5}{120}+\mathcal{O}(\Delta t^6)$$
Al sumar ambas ecuaciones se btiene lo siguiente:
$$y(t+\Delta t)+y(t-\Delta t)=2y(t)+y''(t)\Delta t^2+y''''(t)\frac{\Delta t^4}{12}+\mathcal{O}(\Delta t^6)$$
Al convertir la función continua en una sucesión tal que $y(t)=y_n$ y $\Delta t=h$ entonces se tiene lo siguiente:
$$y_{n+1}+y_{n-1}=2y_n+y_n''h^2+y_n''''\frac{h^4}{12}+\mathcal{O}(h^6)$$
Por otro lado se tiene la siguiente ecuación diferencial:
$$\frac{d^2y}{dx^2}-R(x)y=S(x)$$
Al convertir esta ecuación diferencial de una función continua en una sucesión se tiene lo siguiente:
$$y_n''-R_n y_n=S_n\quad \quad \quad y_n''=R_n y_n+S_n$$
Por otro lado, se puede hallar la segunda derivada de una sucesión de la siguiente manera:
$$f''_n=\frac{f_{n+1}-2f_{n}+f_{n-1}}{h^2}$$
Etonces, se puede obtener la cuarta derivada de la ecuación diferencial de la siguiente manera:
$$y''''_n=\frac{R_{n+1} y_{n+1}+S_{n+1}-2R_n y_n-2S_n+R_{n-1} y_{n-1}+S_{n-1}}{h^2}$$
Al reemplazar la segunda derivadas y la cuarta derivada en la fórmula de orden 6 se pobtiene lo siguiente:
$$y_{n+1}+y_{n-1}=2y_n+\left(R_n y_n+S_n\right)h^2+\left(\frac{R_{n+1} y_{n+1}+S_{n+1}-2R_n y_n-2S_n+R_{n-1} y_{n-1}+S_{n-1}}{h^2}\right)\frac{h^4}{12}+\mathcal{O}(h^6)$$
Se simplifica:
$$y_{n+1}+y_{n-1}=2y_n+\left(R_n y_n+S_n\right)h^2+\left(\frac{R_{n+1} y_{n+1}+S_{n+1}-2R_n y_n-2S_n+R_{n-1} y_{n-1}+S_{n-1}}{12}\right)h^2+\mathcal{O}(h^6)$$
$$y_{n+1}+y_{n-1}=2y_n+\left(R_n y_n+S_n+\frac{R_{n+1} y_{n+1}+S_{n+1}-2R_n y_n-2S_n+R_{n-1} y_{n-1}+S_{n-1}}{12}\right)h^2+\mathcal{O}(h^6)$$
$$y_{n+1}+y_{n-1}=2y_n+\left(\frac{R_{n+1} y_{n+1}+S_{n+1}+10R_n y_n+10S_n+R_{n-1} y_{n-1}+S_{n-1}}{12}\right)h^2+\mathcal{O}(h^6)$$
$$y_{n+1}+y_{n-1}=2y_n+(R_{n+1}y_{n+1}+10R_{n}y_{n}+R_{n-1}y_{n-1})\frac{h^2}{12}+(S_{n+1}+10S_n+S_{n-1})\frac{h^2}{12}+\mathcal{O}(h^6)$$
$$y_{n+1}+y_{n-1}-2y_n-(R_{n+1}y_{n+1}+10R_{n}y_{n}+R_{n-1}y_{n-1})\frac{h^2}{12}=(S_{n+1}+10S_n+S_{n-1})\frac{h^2}{12}+\mathcal{O}(h^6)$$
$$\left(1-R_{n+1}\frac{h^2}{12}\right)y_{n+1}+\left(-2-10R_{n}\frac{h^2}{12}\right)y_n+\left(1-R_{n-1}\frac{h^2}{12}\right)y_{n-1}=(S_{n+1}+10S_n+S_{n-1})\frac{h^2}{12}+\mathcal{O}(h^6)$$
$$\left(1-R_{n+1}\frac{h^2}{12}\right)y_{n+1}-\left(2+R_{n}\frac{5h^2}{6}\right)y_n+\left(1-R_{n-1}\frac{h^2}{12}\right)y_{n-1}=(S_{n+1}+10S_n+S_{n-1})\frac{h^2}{12}+\mathcal{O}(h^6)$$
Y así se demuestra el Algoritmo de Numerov

b La ecuación de Schrodinger es:
$$\frac{-\hbar^2}{2m}\frac{d^2\psi}{dx^2}+V(x)\psi=E(x)\psi$$
Aquí $\psi$ es la función a buscar y $x$ es la variable independiente.

Se pasa el producto de la funcion de onda y la energía a restar:
$$\frac{-\hbar^2}{2m}\frac{d^2\psi}{dx^2}+V(x)\psi-E(x)\psi=0$$
$$\frac{-\hbar^2}{2m}\frac{d^2\psi}{dx^2}+\left(V(x)-E(x)\right)\psi=0$$
Y se divide la ecuación por $\frac{-\hbar^2}{2m}$:
$$\frac{d^2\psi}{dx^2}-\frac{2m}{\hbar^2}\left(V(x)-E(x)\right)\psi=0$$
Al comparar esta ecuación con la ecuación diferencial del punto anterior:
$$\frac{d^2y}{dx^2}-R(x)y=S(x)$$
Se puede evidenciar que:
$$R(x)=\frac{2m}{\hbar^2}\left(V(x)-E(x)\right)\quad\quad\quad\quad S(x)=0$$
Para los estados estacionarios del oscilador armónico, la energía total es un número semientero independiente de $x$:
$$E=\hbar\omega\left(n+\frac{1}{2}\right)$$
Y la energía potencial es:
$$V(x)=\frac{m\omega^2 x^2}{2}$$
Al reemplazar esto en la ecuación anterior se obtiene lo siguiente:
$$R(x)=\frac{2m}{\hbar^2}\left(\frac{m\omega^2 x^2}{2}-\hbar\omega\left(n+\frac{1}{2}\right)\right)\quad\quad\quad\quad S(x)=0$$
Si se asue que $m=1$, $\hbar=1$ y $\omega=1$, entonces las ecuaciones se reducen a:
$$R(x)=x^2-(2n+1)\quad\quad\quad\quad S(x)=0$$
